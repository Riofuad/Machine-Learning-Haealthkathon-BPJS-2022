## Handling Outliers

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_clean.describe()
```

```python
def detect_outliers(df, x):
    Q1 = df[x].describe()['25%']
    Q3 = df[x].describe()['75%']
    IQR = Q3-Q1
    return df[(df[x] < Q1-1.5*IQR) | (df[x] > Q3+1.5*IQR)]
```

### dati2

```python
plt.plot(figsize = (10, 5))
sns.boxplot(df_clean['dati2'])
plt.annotate('Outlier', (df_clean['dati2'].describe()['max'],0.1), xytext = (df_clean['dati2'].describe()['max'],0.3),
             arrowprops = dict(facecolor = 'blue'), fontsize = 13 )
IQR = df_clean['dati2'].describe()['75%'] - df_clean['dati2'].describe()['25%']
```
